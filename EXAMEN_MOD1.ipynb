```python
class TiendaOnline:
    def __init__(self,nombre):
        self.nombre = nombre
        self.inventario = [] #lista de diccionarios {} para almacenar los productos
        self.ventas_totales = 0.0 #float
        
#-----------1---------
    def agregar_producto(self,nombre,precio, cantidad): # función para agregar/revisar productos al inventario
        for producto in self.inventario: #por cada (x) producto que este en el inventario
            if producto ["nombre"]==nombre: #si el nombre del (x) producto es igual a uno que esté ya agregado
                print(f"El producto ({nombre}) existe") #muestramelodime que ya existe

        
                producto ["cantidad"] += cantidad #si ya existe, actualiza la cantidad
                print( f"El inventario a sido actualizado, se agregó {nombre}") #dime que ya fue actualizado
                return self.inventario  #cierro la función con un return
            
            
        self.inventario.append({"nombre":nombre, "precio": precio, "cantidad":cantidad}) #añade al inventario el nombre (clave:Valor) de cada elemento solicitado
        print(f"El producto {nombre} ha sido agregado") #muestrame el nombre del producto que se agregó
        return self.inventario #cierro la función con un return nuevamente por sintaxis
    
#-----------2---------   
    def  ver_inventario(self): # función para ver los productos en el diccionario/inventario
        for producto in self.inventario: #por cada producto en inventario
            print(f"Nombre: {producto["nombre"]},Precio:${producto["precio"]},Cantidad: {producto["cantidad"]}") #entra al diccionario y muestramelos
            pass #coloqué return self.inventario pero me imprimia 2 veces uno de los productos 
    
#-----------3---------          
    def buscar_producto(self, nombre): #función para buscar un producto y cuando lo encuentre te de el detalle de información
        for producto in self.inventario:  #por cada (x) producto que este en el inventario
            if producto ["nombre"]==nombre: #reutilicé logica del ---1--- ya que entra a revisar cada producto
                print(f"Nombre: {producto["nombre"]},Precio:${producto["precio"]},Cantidad: {producto["cantidad"]}") #print rehutilizado del --2--
                return #tenia aca self.inventario tambien pero lo quité porque me daba la inf del producto y tambein el inventario completo
                
#-----------4---------
    def actualizar_stock(self, nombre, cantidad): #función para actualizar el stock 
        for producto in self.inventario: #por cada (x) producto que este en el inventario
            if producto ["nombre"]==nombre: #si el nombre del (x) producto es igual a uno que esté ya agregado
                producto ["cantidad"] += cantidad #si ya existe, actualiza la cantidad 
                print("Inventario actualizado") #indicame que ha sido actualizado el inventario
                print(f"Nombre: {producto["nombre"]},Precio:${producto["precio"]},Cantidad: {producto["cantidad"]}") #traeme el inventario completo 
                return 

#-----------5---------
            
    def eliminar_producto(self, nombre):
        for producto in self.inventario: #por cada (x) producto que este en el inventario
            if producto ["nombre"]==nombre: #si el nombre del (x) producto es igual a uno que esté ya agregado
                self.inventario.pop(nombre)
                print(f"el producto {nombre} ha sido eliminado")
                return

```
