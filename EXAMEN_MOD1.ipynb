```python
class TiendaOnline:
    def __init__(self,nombre):
        self.nombre = nombre
        self.inventario = [] #lista de diccionarios {} para almacenar los productos
        self.ventas_totales = 0.0 
        #---Bonus---
        self.cliente= {} #Diccionario

#-----------1---------
    def agregar_producto(self,nombre,precio, cantidad): # función para agregar/revisar productos al inventario
        for producto in self.inventario: #por cada (x) producto que este en el inventario
            if producto ["nombre"]==nombre: #verifica si el producto "insertado" es igual a uno que esté ya agregado en el inventario
                producto ["cantidad"] += cantidad #si el producto por el cual "consultas" existe, actualiza la cantidad
                print( f"El inventario a sido actualizado, se agregó {nombre}") #dime que ya fue actualizado el producto
                return self.inventario  #cierro la función con un return / opcional :  dejo ver el inventario actualizado 
            
        #si el prpducto no está, agregalo:
        self.inventario.append({"nombre":nombre, "precio": precio, "cantidad":cantidad}) #añade al inventario la {clave:Valor} de cada elemento que se a solicitado 
        print(f"El producto {nombre} ha sido agregado") #muestrame el nombre del producto que se agregó
        return self.inventario #cierro la función con un return / el self es opcional me parece practico ver de inmediato el inventario luego de ser actualizado
    
        #saqué el self para que entrara al bucle a nivel del for registrara el producto en el inventario

#-----------2---------   
    def  ver_inventario(self): # función para ver los productos en el diccionario/inventario
        for producto in self.inventario: #por cada producto en inventario
            print(f"Nombre: {producto["nombre"]},Precio:${producto["precio"]},Cantidad: {producto["cantidad"]}") #entra al diccionario y muestramelos
            continue #gregué pass para que no rompiera el flujo, pero finalmente lo cambié a continue
    
#-----------3---------          
    def buscar_producto(self, nombre): #función para buscar un producto y cuando lo encuentre te de el detalle de su información
        for producto in self.inventario:  #por cada (x) producto que este en el inventario
            if producto ["nombre"]==nombre: #verifica si el producto "insertado" es igual a uno que esté ya agregado en el inventario/ reutilicé logica del ---1--- 
                print(f"Nombre: {producto["nombre"]},Precio:${producto["precio"]},Cantidad: {producto["cantidad"]}") #entra al diccionario y muestramelos/ print rehutilizado del --2--
                return 
                
#-----------4---------
    def actualizar_stock(self, nombre, cantidad): #función para actualizar el stock (agregar /quitar)
        for producto in self.inventario: #por cada (x) producto que este en el inventario
            if producto ["nombre"]==nombre: #si el nombre del (x) producto es igual a uno que esté ya agregado
                producto ["cantidad"] += cantidad #si ya existe, actualiza la cantidad 
                print("Inventario actualizado") #indicame que ha sido actualizado el inventario
                print(f"Nombre: {producto["nombre"]},Precio:${producto["precio"]},Cantidad: {producto["cantidad"]}") #traeme los datos del producto actualizado
                return 

#-----------5---------       
    def eliminar_producto(self, nombre): #función para eliminar productos que ya no queremos vender
        for producto in self.inventario: #por cada (x) producto que este en el inventario
            if producto ["nombre"]==nombre: #si el nombre del (x) producto es igual a uno que esté ya agregado
                self.inventario.remove(producto) # remuevelo / use un remove ya que es el que elimina en diccionario
                print(f"el producto {nombre} ha sido eliminado") #muestrame el producto que se eliminó
                break #intenté con pass/retur pero me imprimia ambos, pense que podia ser la indexación del else probé pero finalmente que obté por break
            else:
                print("Ese producto ingresado no existe en el inventario") 
                return #cierro la función con un return

#-----------6---------

    def calcular_valor_inventario(self): #función para calcular el valor del inventario actual
        ganancia = 0 #variable para guardar monto de patrominio
        ecuacion_producto=0 #variable que guardará la ecuación que necesito hacer

        for producto in self.inventario: #por cada (x) producto que este en el inventario
            ecuacion_producto = producto ["precio"] * producto ["cantidad"] # la ecuación almacenará el resultado de  precio * cantidad
            ganancia += ecuacion_producto # ganancia se alimentará de la ecuación que realice ecuación_producto 
        print(f"El patromonio actual de la empresa es: {ganancia}$") # sacar el print al for para que me lo muestre despues de haber incluido todos los productos
        return 

    
#***********-BONUS-***********

#-----------1--------------

    def agregar_cliente(self, nombre, email): #función para agregar clientes a la empresa
        if nombre not in self.cliente: # si el (x) cliente no estaá en BBDD
            self.cliente[nombre] = {'email':email,'compras': []} # agrega su nombre, email y cantidad el compras (veces o dinero ?)
            print(f"el cliente {nombre} ha sido agregado correctamente") # muestrame el detalle completo del cliente que hemos agregado
            return

        else: 
            print(f"el cliente {nombre} ya existía en la base de datos") #si el cliente ya existia avisamé 
             
#-----------2--------------
    def ver_clientes(self): # función para ver a los clientes
        for clave,valor in self.cliente.items(): #por cada (x) {clave:valor} que esté en la BBDD devuelveme la clave (nombre del cliente) y su valor asociado
            print(f"Nombre: {clave}, Email: {valor['email']}, Compras: {len(valor['compras'])}") #imprime la BBDD de clientes
            continue
```
